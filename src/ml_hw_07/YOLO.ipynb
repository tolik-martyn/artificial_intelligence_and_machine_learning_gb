```python
from ultralytics import YOLO
```

```python
# –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏ –Ω–∞ –Ω–∞–±–æ—Ä–µ –¥–∞–Ω–Ω—ã—Ö COCO128 –≤ —Ç–µ—á–µ–Ω–∏–µ 10 —ç–ø–æ—Ö
model.train(data='coco128.yaml', epochs=10, batch=16, lrf=00.1, lr0=0.001)
```
